```python
#importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#uploading the data
df = pd.read_csv("final_cleaned_sleep_data_v2.csv")

#creating the boxplot itself, making axes and specifying the required variables
plt.figure(figsize=(14, 7))
sns.boxplot(
    data=df,
    x="occupation",
    y="sleep_quality_score",
    hue="work_schedule"
)

#naming the graph and the axes
plt.title("Sleep Quality Score by Occupation and Work Schedule", fontsize=14)
plt.xlabel("Occupation")
plt.ylabel("Sleep Quality Score")


plt.xticks(rotation=45, ha="right")
plt.legend(title="Work Schedule")

plt.tight_layout()
plt.show()
```
