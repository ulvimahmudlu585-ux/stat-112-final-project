```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Loading the data
df = pd.read_csv('final_cleaned_sleep_data_v2.csv')

# 2. Seting the plot theme
sns.set_theme(style="whitegrid")

# 3. Creatimg a scatter plot with regression lines (lmplot) grouped by gender
# hue='gender' -> Separate colors by gender
# palette='magma' -> Color palette (options: 'viridis', 'muted', 'Set1')
g = sns.lmplot(data=df,
               x='stress_score',
               y='sleep_quality_score',
               hue='gender',
               palette='magma',
               height=6,
               aspect=1.5,
               scatter_kws={'alpha':0.5})

# 4. Adjusting the titles and labels
g.fig.suptitle('Correlation between Stress Score and Sleep Quality by Gender', fontsize=15)
g.set_axis_labels('Stress Score', 'Sleep Quality Score')

# Adjusting the layout to prevent the title from overlapping with the plot
plt.subplots_adjust(top=0.9)

plt.show()
```
